```python
# Librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Se declaran las variables de las columnas
columnas = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration",
            "model_year", "origin", "car_name"]
```

```python
# Carga y limpieza de datos
df = pd.read_csv("./auto-mpg/auto-mpg.data", delim_whitespace=True, names=columnas, na_values='?')

df.dropna(inplace=True)
df.drop("car_name", axis=1, inplace=True)
```

```python
# Matriz de correlacion
correlacion = df.corr(numeric_only=True)

print("Matriz de correlacion:")
display(correlacion)

correlacion_mpg = correlacion["mpg"].sort_values(ascending=False)

display(correlacion_mpg)
```

```python
# Se declaran variables dependientes y la independiente
X = df[["displacement", "cylinders", "horsepower", "weight", "acceleration", "model_year"]]

y = df["mpg"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
mlp = MLPRegressor(hidden_layer_sizes=(8,), max_iter=3000, activation='relu',
                   solver='adam', random_state=42, verbose=True)
```

```python
mlp.fit(X_train, y_train)
```

```python
y_pred = mlp.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Error cuadrado medio: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")
```
